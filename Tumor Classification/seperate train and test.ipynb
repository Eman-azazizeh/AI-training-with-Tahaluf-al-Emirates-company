```python
from PIL import Image
import os
import shutil
import pandas as pd

# Path to the directory containing your image files
image_directory = 'C:\\Users\\GTS\\Desktop\\AI Fatima\\Competitions\\4.Tumer classification\\images'

# Load your test DataFrame
test = pd.read_csv("C:\\Users\\GTS\\Desktop\\AI Fatima\\Competitions\\4.Tumer classification\\testset.csv", index_col=0)  # Adjust the path and format as needed
train = pd.read_csv("C:\\Users\\GTS\\Desktop\\AI Fatima\\Competitions\\4.Tumer classification\\trainset.csv", index_col=0)  # Adjust the path and format as needed
```

```python
# Path to the output directory for the Test set
test_output_directory = 'C:\\Users\\GTS\\Desktop\\AI Fatima\\Competitions\\4.Tumer classification\\test'
train_output_directory = 'C:\\Users\\GTS\\Desktop\\AI Fatima\\Competitions\\4.Tumer classification\\train'

# Get a list of all image files in the directory
image_files = [f for f in os.listdir(image_directory) if f.endswith('.png')]

# Loop through the image files
for image_file in image_files:
    image_id = os.path.splitext(image_file)[0]  # Extract the ID from the filename
    print(image_id)
    
    # Check if the image_id is in the test DataFrame
    if int(image_id) in test.iloc[:, 0].values:
        source_path = os.path.join(image_directory, image_file)
        destination_path = os.path.join(test_output_directory, image_file)
        shutil.move(source_path, destination_path)
        
    if int(image_id) in train.iloc[:, 0].values:
        source_path = os.path.join(image_directory, image_file)
        destination_path = os.path.join(train_output_directory, image_file)
        shutil.move(source_path, destination_path)
```
